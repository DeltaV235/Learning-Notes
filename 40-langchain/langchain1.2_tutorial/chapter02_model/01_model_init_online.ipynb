```python
import os

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_deepseek import ChatDeepSeek
from langchain_openai import ChatOpenAI

load_dotenv(override=True)

DEEPSEEK_API_BASE = os.getenv("DEEPSEEK_API_BASE")
DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")

chat = ChatDeepSeek(base_url=DEEPSEEK_API_BASE,
                    api_key=DEEPSEEK_API_KEY,
                    model="deepseek-v4-flash")
chat = ChatDeepSeek(model="deepseek-v4-flash")
response = chat.invoke(input="hello, 你是什么模型")
print(response)
```

## OpenAI 兼容

```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os

load_dotenv(override=True)
DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_API_BASE = os.getenv("DEEPSEEK_API_BASE")
model = ChatOpenAI(model="deepseek-v4-flash", api_key=DEEPSEEK_API_KEY, base_url=DEEPSEEK_API_BASE, )
response = model.invoke("2.5 * 100 = ?")
print(response)
```
