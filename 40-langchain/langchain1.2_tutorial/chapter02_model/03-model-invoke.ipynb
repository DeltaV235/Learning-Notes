# Model Invoke

## invoke

### 字典列表

```python
from dotenv import load_dotenv
from langchain.chat_models.base import init_chat_model
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv(override=True)

model = init_chat_model(
    model="deepseek:deepseek-v4-flash",
)
message = [
    {"role": "system", "content": "你是一位 Python 专家，解释用户提出的关于 Python 的问题"},
    {"role": "user", "content": "... 是什么语法"}
]
response = model.invoke(message)
print(response.content)
message.append({"role": "assistant", "content": f"{response.content}"})
message.append({"role": "user", "content": "用户刚在问了什么"})
response = model.invoke(message)
print(response.content)
```

### 消息对象列表

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

message = [
    SystemMessage(content="你是一位 Python 专家，解释用户提出的关于 Python 的问题"),
    HumanMessage(content="... 是什么语法"),
]
response = model.invoke(message)
print(response.content)

message.append(AIMessage(content=response.content))
message.append(HumanMessage(content="用户刚才问了什么"))
response = model.invoke(message)
print(response.content)
```


## AIMessage

```python
model = init_chat_model(model="openrouter:openai/gpt-5.6-luna-pro")
response = model.invoke("hello")
print(type(response))
```

```python
from rich import print as rprint
rprint(response)
```

```python
# 1. 获取回复内容
print("AI 回复:", response.content)
# 2. 获取响应元数据
metadata = response.response_metadata
print(f"使用的模型: {metadata.get('model_name')}")
print(f"结束原因: {metadata['finish_reason']}")
print(f"模型提供商：{metadata['model_provider']}\n")
# 3. 获取 Token 使用情况
usage = metadata.get('token_usage', {})
print(f"输入 tokens: {usage.get('prompt_tokens')}")
print(f"输出 tokens: {usage.get('completion_tokens')}")
print(f"总计 tokens: {usage.get('total_tokens')}")
# 4. 获取消息 ID
print(f"消息 ID: {response.id}")
```
