# Model Stream & Batch Invocation

## Stream

```python
from dotenv import load_dotenv
from langchain.chat_models.base import init_chat_model
from rich import print as rprint

load_dotenv(override=True)

model = init_chat_model(
    model="deepseek:deepseek-v4-flash",
)
message = [
    {"role": "system", "content": "你是一位 Python 专家，解释用户提出的关于 Python 的问题"},
    {"role": "user", "content": "... 是什么语法"}
]
```

```python
for chunk in model.stream(input=message):
    print(chunk.text, end="", flush=True)
```

## Batch

```python
messages = [
    "你好，你是谁？",
    "2 + 3 * 5 = ?",
    "中国首都在哪里？"
]
responses = model.batch(inputs=messages)

rprint(responses)
```

```python
responses = model.batch_as_completed(inputs=messages)

for response in responses:
    rprint(response)
```
